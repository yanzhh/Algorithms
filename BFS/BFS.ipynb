Define a graph by Adjacent list

we want to do a Breadth-First-Search and find the levels and parents of every point.
We use dictionary to store levels and parents information.

```python
Adj = {'s':['a','x'], 'a':['s','z'],'z':['a'], 'x':['s','d','c'], 'd':['f','c','x'], 'c':['x','d','f','v'], 'f':['d','c','v'], 'v':['c','f']}
```

```python
def BFS(Adj,s):
    """
    Input:
    Adj: Adjacent list, represent a graph.
    s: start point
    Output:
    level: dictionary
    parent: dictionary
    """
    level = {s:0}   #start point is in current level: level 0
    parent = {s:None}
    i = 1 # next level is i
    frontier = [s]  #points in previous level: level i-1
    while len(frontier)!= 0: #level
        next = []   # points in next level
        for u in frontier:  #levelpoints
            for v in Adj[u]:    #uneighbours.
                if v not in level.keys():    #vlevel
                    next.append(v)  #nextv
                    level[v] = i    #vleveli
                    parent[v] = u   #vparentu
        frontier = next
        i += 1
    return level, parent
```

```python
level, parent = BFS(Adj,'s')
print('level',level)
print('parent',parent)
```
